```python
import seaborn as sns
iris = sns.load_dataset('iris')
```

```python
virginica_sepal_width = iris[iris['species'] == 'virginica']['sepal_width']
min_sepal_width_virginica = virginica_sepal_width.min()
print(f"The minimum sepal_width for the 'virginica' variety is: {min_sepal_width_virginica}")
```

```python
mean_petal_length = iris['petal_length'].mean()
filtered_iris = iris[iris['petal_length'] > mean_petal_length]
display(f"The mean petal_length of the entire iris dataset is: {mean_petal_length}")
display(filtered_iris.head())
```

```python
num_observations = len(filtered_iris)
display(f"The number of observations in the new DataFrame (where petal_length > mean_petal_length) is: {num_observations}")
```

```python
correlation_sepal_length_sepal_width = iris['sepal_length'].corr(iris['sepal_width'])
print(f"Pearson correlation between sepal_length and sepal_width: {correlation_sepal_length_sepal_width:.4f}")
```

```python
correlation_petal_length_sepal_length = iris['petal_length'].corr(iris['sepal_length'])
print(f"Pearson correlation between petal_length and sepal_length: {correlation_petal_length_sepal_length:.4f}")
```

```python
correlation_sepal_length_petal_width = iris['sepal_length'].corr(iris['petal_width'])
print(f"Pearson correlation between sepal_length and petal_width: {correlation_sepal_length_petal_width:.4f}")
```

```python
correlation_petal_length_petal_width = iris['petal_length'].corr(iris['petal_width'])
print(f"Pearson correlation between petal_length and petal_width: {correlation_petal_length_petal_width:.4f}")
```

```python
average_sepal_width_by_species = iris.groupby('species')['sepal_width'].mean()
display(f"Average sepal_width by species:\n{average_sepal_width_by_species}")
```

```python
# Mean petal_length of setosa
setosa_petal_length_mean = iris[iris['species'] == 'setosa']['petal_length'].mean()
display(f"Mean petal_length of setosa: {setosa_petal_length_mean:.4f}")

# Iris with the largest sepal_width
largest_sepal_width_iris = iris.loc[iris['sepal_width'].idxmax()]
display(f"Iris with the largest sepal_width:\n{largest_sepal_width_iris}")

# Standard deviation of sepal_length of setosa
setosa_sepal_length_std = iris[iris['species'] == 'setosa']['sepal_length'].std()
display(f"Standard deviation of sepal_length of setosa: {setosa_sepal_length_std:.4f}")

# Iris with the largest sepal_length
largest_sepal_length_iris = iris.loc[iris['sepal_length'].idxmax()]
display(f"Iris with the largest sepal_length:\n{largest_sepal_length_iris}")
```

```python
mean_sepal_length_rounded = iris['sepal_length'].mean()
num_unique_species = iris['species'].nunique()
num_virginica = len(iris[iris['species'] == 'virginica'])

display(f"Mean sepal_length (rounded to 3 digits): {mean_sepal_length_rounded:.3f}")
display(f"Number of unique species: {num_unique_species}")
display(f"Number of 'virginica' species: {num_virginica}")
```

```python
num_versicolor = len(iris[iris['species'] == 'versicolor'])
display(f"The number of observations for 'versicolor' species is: {num_versicolor}")
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

```python
print('Missing values in Iris dataset:')
missing_iris = iris.isnull().sum()
display(missing_iris[missing_iris > 0])

print('\nMissing values in Titanic dataset:')
missing_titanic = titanic.isnull().sum()
display(missing_titanic[missing_titanic > 0])
```

```python
first_class_not_survived_ages = titanic[(titanic['pclass'] == 1) & (titanic['survived'] == 0)]['age']
min_age_first_class_not_survived = first_class_not_survived_ages.min()
max_age_first_class_not_survived = first_class_not_survived_ages.max()

display(f"The minimum age for 1st class passengers who did not survive is: {min_age_first_class_not_survived}")
display(f"The maximum age for 1st class passengers who did not survive is: {max_age_first_class_not_survived}")
```

```python
female_passengers = titanic[titanic['sex'] == 'female']
num_female_survived = female_passengers['survived'].sum()
total_female_passengers = len(female_passengers)

if total_female_passengers > 0:
    percentage_female_survived = (num_female_survived / total_female_passengers) * 100
    rounded_percentage = round(percentage_female_survived)
    display(f"The percentage of female passengers who survived is: {rounded_percentage:.0f}%")
else:
    display("There are no female passengers in the dataset.")
```

```python
survived_second_class_passengers = titanic[(titanic['pclass'] == 2) & (titanic['survived'] == 1)]
num_survived_second_class = len(survived_second_class_passengers)
display(f"The number of passengers in the 2nd class who survived is: {num_survived_second_class}")
```

## Correlation with Passenger Class

### Subtask:
Calculate the correlation between 'survived' and 'pclass' to verify the fourth statement. Note that a lower 'pclass' value indicates a higher class.


**Reasoning**:
To calculate the correlation between 'survived' and 'pclass', I will use the `.corr()` method on the `titanic` DataFrame.



```python
correlation_survived_pclass = titanic['survived'].corr(titanic['pclass'])
print(f"Pearson correlation between 'survived' and 'pclass': {correlation_survived_pclass:.4f}")
```

## Correlation with Fare

### Subtask:
Calculate the correlation between 'survived' and 'fare' to further analyze factors affecting survival rates.

**Reasoning**:
To calculate the correlation between 'survived' and 'fare', I will use the `.corr()` method on the `titanic` DataFrame.



```python
correlation_survived_fare = titanic['survived'].corr(titanic['fare'])
print(f"Pearson correlation between 'survived' and 'fare': {correlation_survived_fare:.4f}")
```
